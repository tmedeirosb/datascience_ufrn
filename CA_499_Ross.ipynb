```python
#exemplo livro ross pag. 499
import pandas as pd
import prince
import numpy as np

columns = ['Democrat', 'Republican', 'Independent']
rows = ['Women', 'Men']
values = [
    [68, 56, 32],
    [52, 72, 20],
]

df = pd.DataFrame(data=values, index=rows, columns=columns)

df
```

```python
#calculo da analise de correspondencia
ca = prince.CA(df, n_components=2)
```

```python
#soma das linhas
np.diag(ca.row_sums * ca.N)
```

```python
#soma das colunas
np.diag(ca.column_sums * ca.N)
```

```python
#frequencia 
ca.P
```

```python
#soma de todas as frequencias dá 1.0
np.sum(ca.P.values)
```

```python
#normalizado por linha
print(ca.row_profiles)
print(ca.row_profiles.iloc[0,:].sum())
print(ca.row_profiles.iloc[1,:].sum())
ca.row_profiles.sum(axis='columns')
```

```python
#normalizado por coluna
print(ca.column_profiles)
ca.column_profiles.sum(axis='rows')
```

```python
#frequencia de cada valor da linha
ca.row_sums
```

```python
#frequencia de cada valor por coluna
ca.column_sums
```

```python
#frequencia esperada N*M/n
ca.expected_frequencies * ca.N
```

```python
#frequencia relativa
ca.P - ca.expected_frequencies
```

```python
#residual matrix
(ca.P - ca.expected_frequencies) * ca.N
```

```python
#explained inertial
ca.explained_inertia
```

```python
np.sqrt(np.diag(ca.row_sums)) @ ca.svd.U
```

```python
((68-62.40)**2)/62.40
```

```python
freq_esperada = ca.expected_frequencies * ca.N
freq_esperada
```

```python
qui_quadrado = (freq_esperada - df)**2/df
qui_quadrado.values.sum()
```
